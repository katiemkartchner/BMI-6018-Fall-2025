```python
#create new tables of DIAGNOSIS and DRUG_ORDER

#DIAGNOSIS table
# --- CREATE TABLES ---
cursor.executescript("""
CREATE TABLE DIAGNOSIS (
    DxID INTEGER PRIMARY KEY,
    PID INTEGER,
    DxCode TEXT,
    InitialDxDts DATE,
    DxStatus TEXT,
    FOREIGN KEY (PID) REFERENCES PATIENT(PID),
    FOREIGN KEY (DxCode) REFERENCES ICD(Code)
);

CREATE TABLE DRUG_ORDER (
    OrderID INTEGER PRIMARY KEY,
    PID INTEGER,
    DID INTEGER,
    OrderDts DATE,
    OrderStatus TEXT,
    FOREIGN KEY (PID) REFERENCES PATIENT(PID),
    FOREIGN KEY (DID) REFERENCES DRUG_FORMULARY(DID)
);
""")

```

```python
#4. List drug orders (DID) having OrderStatus with either Complete or Pending.
cursor.execute("""
SELECT DID, OrderStatus 
FROM DRUG_ORDER 
WHERE OrderStatus IN ('Complete', 'Pending');
""")
print("4. Drug orders with Complete or Pending:")
for row in cursor.fetchall():
    print(row)
```

```python
#5. List ICD diagnosis codes and names with their effective dates earlier than 11/2/2017.
cursor.execute("""
SELECT Code, Description, EffectiveDts
FROM ICD
WHERE EffectiveDts < '2017-11-02';
""")
print("5. ICD codes before 2017-11-02:")
for row in cursor.fetchall():
    print(row)
```

```python
#6. List ICD diagnosis codes and names having word “Malignant” in description.
cursor.execute("""
SELECT Code, Description
FROM ICD
WHERE Description LIKE '%Malignant%';
""")
print("6. ICDs with 'Malignant' in description:")
for row in cursor.fetchall():
    print(row)
```

```python
#7. For each drug, show how many times they were ordered after 2010. (on or after 2011.1.1).
cursor.execute("""
SELECT DID, COUNT(*) AS OrderCount
FROM DRUG_ORDER
WHERE OrderDts >= '2011-01-01'
GROUP BY DID;
""")
print("7. Orders per drug after 2010:")
for row in cursor.fetchall():
    print(row)
```

```python
#8. List medication formularies with their dose form and route are null.
cursor.execute("""
SELECT Name
FROM DRUG_FORMULARY
WHERE DoseForm IS NULL AND Route IS NULL;
""")
print("8. Drugs with DoseForm and Route NULL:")
for row in cursor.fetchall():
    print(row)
```

```python
#9. List first and last names of patients having at least one diagnosis.
cursor.execute("""
SELECT DISTINCT P.FirstName, P.LastName
FROM PATIENT P
JOIN DIAGNOSIS D ON P.PID = D.PID;
""")
print("9. Patients with at least one diagnosis:")
for row in cursor.fetchall():
    print(row)
```

```python
#10. List first and last names of patients having both diagnosis and medication (at least once for each).
cursor.execute("""
SELECT DISTINCT P.FirstName, P.LastName
FROM PATIENT P
JOIN DIAGNOSIS D ON P.PID = D.PID
JOIN DRUG_ORDER O ON P.PID = O.PID;
""")
print("10. Patients with both diagnosis and medication:")
for row in cursor.fetchall():
    print(row)
```

```python
#11. List first and last names of patients whose medication order dose form is “Tablet”.
cursor.execute("""
SELECT DISTINCT P.FirstName, P.LastName
FROM PATIENT P
JOIN DRUG_ORDER O ON P.PID = O.PID
JOIN DRUG_FORMULARY F ON O.DID = F.DID
WHERE F.DoseForm = 'Tablet';
""")
print("11. Patients whose medication dose form is Tablet:")
for row in cursor.fetchall():
    print(row)
```

```python
#12. List names and codes of medication ordered at least twice.
cursor.execute("""
SELECT F.Name, O.DID, COUNT(*) AS OrderCount
FROM DRUG_ORDER O
JOIN DRUG_FORMULARY F ON O.DID = F.DID
GROUP BY O.DID
HAVING COUNT(*) >= 2;
""")
print("12. Drugs ordered at least twice:")
for row in cursor.fetchall():
    print(row)
```
